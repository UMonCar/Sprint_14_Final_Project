## 1. Carga y Exploraci√≥n de Datos (Python)

```python
# Cargar todas las librer√≠as
from scipy import stats as st
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
import math as mt
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
import re
import os
from plotly import graph_objects as go
from scipy.stats import shapiro, kstest
#
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# La funci√≥n data_load() carga un archivo CSV desde una carpeta llamada 'datasets' y devuelve un DataFrame con los datos.

def data_load(file = 'telecom_dataset_us.csv'):	
    ruta_archivo = os.path.join('datasets', file)
    data = pd.read_csv(
	    ruta_archivo
	)
    return data
```

```python
# Se llama a la funci√≥n data_loas() y se guarda el datarame resultante en 'clients'
clients = data_load('telecom_clients_us.csv')
# clients.columns = dataset.columns.str.lower()
clients.info()
```

```python
# Se llama a la funci√≥n data_load() y se guarda el datarame resultante en 'dataset'
dataset = data_load('telecom_dataset_us.csv')
# dataset.columns = dataset.columns.str.lower()
dataset.info()
```

```python
# Se convierte la columna 'date' a 'datetime'
# Se agrega la columna 'period' con solo los periodes de tiempo de A√±o - Mes de los datos

dataset['date'] = pd.to_datetime(dataset['date'])
dataset['date'] = dataset['date'].dt.strftime('%Y-%m-%d')
dataset['date'] = pd.to_datetime(dataset['date'])
dataset['period'] = dataset['date'].dt.strftime('%Y-%m')
dataset['period'] = pd.to_datetime(dataset['period'])
```

```python
# Se muestran las primeas 5 columnas del dataframe

dataset.head()
```

```python
# Se eliminan las filas con valores vac√≠os, dado que no es posible identificar el operados o el tipo de llamada.

data = dataset.dropna(subset=['operator_id'])
data = data.dropna(subset=['internal'])

# Se convierte el tipo de la columna 'operator_id' a entero

data['operator_id'] = data['operator_id'].astype(int)
```

```python
# Se calcula el tiempo de espera y se guarda en una nueva columna llamada 'wait time'

data['wait_time'] = data['total_call_duration'] - data['call_duration']
```

Se tomaron los siguientes umbrales para considerar como ineficaces a los operadores:

-   Llamadas salientes menores a: `50`
-   Llamadas perdidas mayores a: `50`
-   Tiempo de espera mayor a: `175`

```python
bad_op_wait_time = wait_time[wait_time['wait_time'] > 175]['operator_id']
bad_op_wait_time = bad_op_wait_time.reset_index(drop= True)
bad_op_missed_calls = missed_calls[missed_calls['is_missed_call'] > 50]['operator_id']
bad_op_missed_calls = bad_op_missed_calls.reset_index(drop= True)
bad_op_out_calls = out_calls[out_calls['direction'] < 50]['operator_id']
bad_op_out_calls = bad_op_out_calls.reset_index(drop= True)
```

2. Calcular la proporci√≥n de llamadas perdidas por operador.
3. Analizar la cantidad de llamadas salientes realizadas por operador.
4. Evaluar el tiempo de espera promedio de cada operador.

```python
bad_operators = pd.concat([bad_op_wait_time, bad_op_missed_calls, bad_op_out_calls], ignore_index= True)
bad_operators = bad_operators.drop_duplicates()
bad_operators.reset_index(drop=True, inplace=True)
print(bad_operators)
```

```python
bad_op_data = data[data['operator_id'].isin(bad_operators)]
good_op_data = data[~ data['operator_id'].isin(bad_operators)]

bad_op_data.info()
good_op_data.info()
```

```python
bad_op_wait_time = wait_time[wait_time['operator_id'].isin(bad_operators)]
bad_op_wait_time = bad_op_wait_time.reset_index(drop= True)
bad_op_missed_calls = missed_calls[missed_calls['operator_id'].isin(bad_operators)]
bad_op_missed_calls = bad_op_missed_calls.reset_index(drop= True)
bad_op_out_calls = out_calls[out_calls['operator_id'].isin(bad_operators)]
bad_op_out_calls = bad_op_out_calls.reset_index(drop= True)
```

```python
bad_op_metrics = bad_op_wait_time.merge(bad_op_missed_calls, on='operator_id', how='outer').merge(bad_op_out_calls, on='operator_id', how='outer')
bad_op_metrics.columns = ['operator_id', 'wait_time', 'missed_calls', 'out_calls']
bad_op_metrics.info()
```

## 4. Pruebas de Hip√≥tesis Estad√≠sticas

```python
data['missed_call_rate'] = data['is_missed_call'] / data['calls_count']

# Segmentar operadores en eficaces e ineficaces (esto depender√° de tus umbrales)
# Ejemplo: si m√°s del 30% de las llamadas fueron perdidas, clasificamos como ineficaz
threshold_missed_calls = 0.30
ineficaces = data[data['missed_call_rate'] > threshold_missed_calls]
eficaces = data[data['missed_call_rate'] <= threshold_missed_calls]
```

```python
print('Tama√±o de los grupos:')
print('Operadores ineficaces: ', len(ineficaces))
print('Operadores eficaces: ', len(eficaces))
```

**Supuesto de Normalidad**
Podemos verificar si los datos de las dos muestras (eficaces e ineficaces) siguen una distribuci√≥n normal, alusar la prueba de Shapiro-Wilk en una muestra grande (m√°s de 5,000 datos), el valor p calculado puede no ser preciso. Para grandes muestras, existen alternativas que funcionan mejor en la verificaci√≥n de normalidad:

Como alternativa para pruebas de normalidad en muestras grandes, se us√≥ la prueba de Anderson-Darling:

Es adecuada para muestras grandes y proporciona una estad√≠stica de bondad de ajuste con varios niveles de significancia.
No entrega un valor p, sino que compara el estad√≠stico con valores cr√≠ticos para determinar si se cumple la normalidad.

```python
from scipy.stats import anderson

# Anderson-Darling Test para eficaces
result_eficaz = anderson(eficaces['missed_call_rate'], dist='norm')
print("Anderson-Darling Test para eficaces - Estad√≠stico:", result_eficaz.statistic)
print("Valores cr√≠ticos:", result_eficaz.critical_values)
print("Niveles de significancia:", result_eficaz.significance_level)

# Anderson-Darling Test para ineficaces
result_ineficaz = anderson(ineficaces['missed_call_rate'], dist='norm')
print("Anderson-Darling Test para ineficaces - Estad√≠stico:", result_ineficaz.statistic)
print("Valores cr√≠ticos:", result_ineficaz.critical_values)
print("Niveles de significancia:", result_ineficaz.significance_level)
```

**Supuesto de Homogeneidad de Varianzas**

Para verificar la homogeneidad de varianzas entre los grupos, puedes utilizar:

Prueba de Levene: Esta prueba eval√∫a si dos o m√°s grupos tienen varianzas iguales. Un valor p menor que 0.05 indica que las varianzas no son homog√©neas.
Prueba de Bartlett: Otra prueba com√∫n, aunque es m√°s sensible a desviaciones de la normalidad.

```python
from scipy.stats import levene, bartlett

# Prueba de Levene para la homogeneidad de varianzas
stat_levene, p_levene = levene(eficaces['missed_call_rate'], ineficaces['missed_call_rate'])
print("Prueba de Levene - Estad√≠stico:", stat_levene, ", p-valor:", p_levene)

# Prueba de Bartlett (si los datos son normales)
stat_bartlett, p_bartlett = bartlett(eficaces['missed_call_rate'], ineficaces['missed_call_rate'])
print("Prueba de Bartlett - Estad√≠stico:", stat_bartlett, ", p-valor:", p_bartlett)
```

Los resultados de las pruebas indican que:

### 1. **Prueba de Anderson-Darling para Normalidad**
   - Ambos grupos (eficaces e ineficaces) tienen un estad√≠stico Anderson-Darling mucho mayor que los valores cr√≠ticos en todos los niveles de significancia (0.576 a 1.092). Esto significa que los datos de ambos grupos **no siguen una distribuci√≥n normal**, dado que los valores exceden ampliamente los umbrales.
  
### 2. **Pruebas de Homogeneidad de Varianzas (Levene y Bartlett)**
   - Los valores p de ambas pruebas (Levene y Bartlett) son 0.0, lo cual es menor que 0.05. Esto indica que **no se cumple el supuesto de homogeneidad de varianzas** entre los dos grupos, lo que significa que las varianzas son significativamente diferentes.

### **Conclusi√≥n y Sugerencias**

Debido a que los datos no cumplen con la normalidad ni con la homogeneidad de varianzas, las pruebas param√©tricas (como la prueba t y ANOVA) no son apropiadas. Las opciones recomendadas son:

1. **Prueba de Mann-Whitney U** (para dos muestras no normales y con varianzas desiguales): Es una prueba no param√©trica que eval√∫a si hay diferencias significativas entre dos muestras independientes.

```python
from scipy.stats import mannwhitneyu

# Prueba de Mann-Whitney U
stat_mw, p_mw = mannwhitneyu(eficaces['missed_call_rate'], ineficaces['missed_call_rate'], alternative='two-sided')
print("Prueba de Mann-Whitney U - Estad√≠stico:", stat_mw, ", p-valor:", p_mw)
```

Un resultado con un **estad√≠stico de Mann-Whitney U de 0** y un **valor p de 0.0** indica una diferencia altamente significativa entre los dos grupos que comparaste (eficaces e ineficaces) en cuanto a la tasa de llamadas perdidas. En este caso:

- **Estad√≠stico U = 0** significa que no hay solapamiento en las distribuciones de los dos grupos. Esto sugiere que, seg√∫n la m√©trica de tasa de llamadas perdidas, los grupos de eficaces e ineficaces son completamente distintos.
- **Valor p = 0.0** (menor a 0.05) indica que la diferencia entre los grupos es estad√≠sticamente significativa, por lo que rechazas la hip√≥tesis nula de que ambos grupos tienen la misma distribuci√≥n en la tasa de llamadas perdidas.

### Interpretaci√≥n
Esto confirma que los operadores clasificados como "ineficaces" tienen una tasa de llamadas perdidas significativamente distinta (m√°s alta) en comparaci√≥n con los operadores "eficaces." Esto respalda la segmentaci√≥n que hiciste para identificar a los operadores menos eficaces y podr√≠a ser una base s√≥lida para generar alertas o acciones de mejora dirigidas a estos operadores en el sistema CallMeMaybe.
